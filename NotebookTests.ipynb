```python
import numpy as np
import IPython.display as ipd
```

```python
input_file = './files/INPUT_AUDIO_ZW1079.txt'
output_file = './files/OUTPUT_AUDIO_ZW1079.txt'
```

```python
def read_file(file):
    with open(file, 'r') as f:
        return f.read()
```

```python
input_data = read_file(input_file)
output_data = read_file(output_file)
```

```python
def string_to_sample(data):
    return np.array(list(map(float, data.split())))
```

```python
ipd.Audio(string_to_sample(output_data), rate=22500)
```
