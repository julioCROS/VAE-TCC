```python
import numpy as np
import IPython.display as ipd
```

```python
audio_id = 'IV2498'
audio_rate = 11000
```

```python
input_file = './results/STG_input_' + str(audio_id) + '.txt'
output_file = './results/STG_generated_2_' + str(audio_id) + '.txt'
```

```python
def read_file(file):
    with open(file, 'r') as f:
        return f.read()
```

```python
input_data = read_file(input_file)
output_data = read_file(output_file)
```

```python
def string_to_sample(data):
    return np.array(list(map(float, data.split())))
```

```python
ipd.Audio(string_to_sample(input_data), rate=audio_rate)
```

```python
ipd.Audio(string_to_sample(output_data), rate=audio_rate)
```
